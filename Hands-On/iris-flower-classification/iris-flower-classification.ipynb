```python
# Standard Imports
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Lets load the dataset
iris_data=pd.read_csv("iris-data.csv")
iris_data
```

```python
# the first step is to get our data ready,our data is all in numeric form so it look up pretty much ready
```

```python
# Now lets select a model for our data, let's try svm
from sklearn import svm

iris_data=iris_data.sample(frac=1)

# Lets split the data into feature and target variables
X=iris_data.drop('species',axis=1)
y=iris_data['species']

# Lets split our data now into training and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

# Lets fit our data to the model
clf= svm.SVC()
clf.fit(X_train,y_train)

# Evaluate our model
clf.score(X_test,y_test)
```

```python
y_preds=clf.predict(X_test)
y_preds
```

```python
from sklearn.metrics import confusion_matrix,classification_report
confusion_matrix(y_test,y_preds)
```

```python
print(classification_report(y_test,y_preds))
```

```python
# Import LinearSVC from sklearn's svm module
from sklearn.svm import LinearSVC

# Import KNeighborsClassifier from sklearn's neighbors module
from sklearn.neighbors import KNeighborsClassifier

# Import SVC from sklearn's svm module
from sklearn.svm import SVC

# Import LogisticRegression from sklearn's linear_model module
from sklearn.linear_model import LogisticRegression

# Import RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

```

```python
models = {"LinearSVC": LinearSVC(),
          "KNN": KNeighborsClassifier(),
          "SVC": SVC(),
          "LogisticRegression": LogisticRegression(),
          "RandomForestClassifier": RandomForestClassifier()}

results={}

for model_name,model in models.items():
    model.fit(X_train,y_train)
    results[model_name]=model.score(X_test,y_test)
        
results
```

```python
max(results.values())
```
