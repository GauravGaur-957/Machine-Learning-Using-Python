# Matplotlib Example Workflow

```python
#0. Import matplotlib and use inline to plot it in your notebook
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare some data to plot
x=[1,2,3,4]
y=[10,11,12,13]

#2. Setup a plot
fig,ax=plt.subplots(figsize=(10,10))  #figsize here symbolises the custom width and height,i.e x and y

#3. Plot the data
ax.plot(x,y)

#4. Customize plot
ax.set(title="Simple Plot",xlabel="x-axis",ylabel="y-axis")

#5. Save and show
fig.savefig("./plots/sample-plot.png")
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100) #generates evenly spaced 100 numbers between 0 to 10
x[:10]
```

```python
#Let's plot the data and create a line plot
fig,ax=plt.subplots()
ax.plot(x,x**2);
```

```python
##Let's plot the data and create a scatter plot this time
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
#Another scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
#Making a plot from a dictionary
prices={'pen':12,'pencil':10,'eraser':8}
fig,ax=plt.subplots()
ax.bar(prices.keys(),height=prices.values());
ax.set(title="Stationery Prices",xlabel='items',ylabel='price');
```

```python
fig,ax=plt.subplots()
ax.barh(list(prices.keys()),list(height=prices.values()));
```

```python
#Make some data for histogram and plot it
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

# 2 options for subplots

```python
#Subplot option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#Plot to different axes
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(prices.keys(),prices.values());
ax4.hist(np.random.randn(1000));
```

```python
#Subplot option 2 (I found this as more cool)
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#Plot to different axes
ax[0][0].plot(x,x/2);
ax[0][1].scatter(np.random.random(10),np.random.random(10));
ax[1][0].bar(prices.keys(),prices.values());
ax[1][1].hist(np.random.randn(1000));
```

## Which plotting method should you use,(pyplot vs matplotlib oop methods)
# When plotting quickly,it is okay to use pyplot method
# When plotting something more advanced, use the OO method

```python
over_50=df[df['age']>50]
over_50.head()
```

```python
#Pyplot Method
over_50.plot(kind='scatter',x='age',y='chol',c='target');
```

```python
#OO Method
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);
ax.set_xlim([45,100]);
```
